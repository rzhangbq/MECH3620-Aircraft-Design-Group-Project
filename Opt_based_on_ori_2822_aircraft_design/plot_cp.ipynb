```python
import glob
import re

import pandas as pd
import matplotlib.pyplot as plt

file_path = "/home/fydp/Documents/SU2/Ray/NACA4412_26_11_23/aoa0_9999/DESIGNS/DSN_*/DIRECT/surface_flow.csv"


# Find all files matching the pattern
file_list = glob.glob(file_path)
file_list.sort()
```

```python
fig, ax = plt.subplots()

# Specify the design numbers to plot
specified_design_numbers = [1, 10, 18, 28, 46, 68]

for file in file_list:
    df = pd.read_csv(file)
    pattern = r'DSN_(\d+)'
    match = re.search(pattern, file)
    design_number = int(match.group(1))
    
    if design_number in specified_design_numbers:
        ax.plot(df['x'], df['Pressure_Coefficient'], label='NACA4412 {}'.format(design_number))

ax.set_xlabel('x/c')
ax.set_ylabel('cp')
ax.invert_yaxis()
ax.set_title('NACA4412 Pressure Distribution')
ax.legend(loc="upper right")

plt.show()
```
