```python
import glob
import re

file_path = "DESIGNS/DSN_*/DIRECT/log_Direct.out"

# Initialize empty arrays to store the extracted numbers
first_numbers = []
second_numbers = []

# Find all files matching the pattern
file_list = glob.glob(file_path)
file_list.sort()
if len(file_list) > 0:
    # Iterate over each file
    for file_name in file_list:
        # Open the file
        with open(file_name, 'r') as file:
            lines = file.readlines()

            # Iterate over the lines in reverse order
            for line in reversed(lines):
                # Check if the line matches the pattern
                if re.match(r'\|\s+\d+\|', line):
                    # Extract the last two numbers
                    extracted_numbers = re.findall(r'-?\d+\.\d+', line)[-2:]

                    # Convert the numbers to float and store them in separate arrays
                    first_numbers.append(float(extracted_numbers[0]))
                    second_numbers.append(float(extracted_numbers[1]))
                    break

    # Print the extracted numbers
    print("First numbers:")
    print(first_numbers)
    print("Second numbers:")
    print(second_numbers)
else:
    print("No matching files found.")
```

```python
import matplotlib.pyplot as plt

# ... (previous code to extract the numbers and store them in first_numbers and second_numbers arrays)

# Create figure and axis objects
fig, ax1 = plt.subplots()

# Create a twin axis for the second_numbers array on the right
ax1.plot(second_numbers, color='red', label='$C_D$')
ax1.set_ylabel('$C_D$')
ax1.tick_params(axis='y')
ax1.set_xlabel('Number of iteration')
# ax1.set_ylim(0, 0.014)
ax1.set_ylim(0.00915, 0.00953)

# Plot the first_numbers array on the left axis
ax2 = ax1.twinx()
ax2.plot(first_numbers, color='blue', label='$C_L$')
ax2.set_xlabel('Index')
ax2.set_ylabel('$C_L$')
ax2.tick_params(axis='y')
ax2.set_ylim(0, 0.8)



# Add legend
lines = ax1.get_lines() + ax2.get_lines()
labels = [line.get_label() for line in lines]
ax1.legend(lines, labels)

# Title
plt.title('NACA4412 Plot of $C_L$ and $C_D$')

# Display the plot
plt.show()
```
